#1.Create DataFrames from scratch part-1.

#DataFrame -> 2D array-2D Matrix
Multicolumn data
*  each column represents different variable
*  each column contains data of its own type
*  we can preserve data consistency(we aim for having information of the same type within a certain column)

**DataFrame is nothing but collection of multiple series objects**

*contains values along two axis(rows and columns)

*An enchanced **python dictonary**

```python
import pandas as pd
```

#1: Construct a DataFrame from a dictonary of lists

keys --> column names
values --> rows

```python
data = {'Product Name':['Product A', 'Product B', 'Product C'], 'Product Price':[22250, 16600, 12500]}
df = pd.DataFrame(data)
df
```

```python
#All arrays must be of the same length
data = {'Product Name':['Product A', 'Product B', 'Product C', 'Product D'], 'Product Price':[22250, 16600, 12500]}
df = pd.DataFrame(data)
df
```

#2: Construct a DataFrame from a dictonary of lists + specify an index

```python
data = {'Product Name':['Product A', 'Product B', 'Product C'], 'Product Price':[22250, 16600, 12500]}

df = pd.DataFrame(data, index = ['A', 'B', 'C'])
df
```

**More professional**

```python
data = {'Product Name':['Product A', 'Product B', 'Product C'], 'Product Price':[22250, 16600, 12500]}
product_IDs = ['A', 'B', 'C']
df = pd.DataFrame(data, index = product_IDs)
df
```

#3: Construct a DataFrame from a list of dictonaries

```python
data = [{'ProductName':'Product A', 'ProductPrice':22250},
        {'ProductName':'Product B', 'ProductPrice':16600},
        {'ProductName':'Product C', 'ProductPrice':12500}]
df = pd.DataFrame(data)
df
```

```python
data = [{'ProductName':'Product A', 'ProductPrice':22250},
        {'ProductName':'Product B', 'ProductPrice':16600},
        {'ProductName':'Product C', 'ProductPrice':[12500, 2000]}]
df = pd.DataFrame(data)
df
```

#4: Construct a DataFrame from a dictionary of pandas Series

```python
ser_products = pd.Series(['Product A', 'Product B', 'Product C'])
ser_prices = pd.Series([22250, 16600, 12500])
```

```python
data = {'Product Name': ser_products, 'Product Prices': ser_prices}
df = pd.DataFrame(data)
df
```

```python
ser_products = pd.Series(['Product A', 'Product B', 'Product C'], index = ['A', 'B', 'C'])
ser_prices = pd.Series([22250, 16600, 12500], index = ['A', 'B', 'C'])
data = {'Product Name': ser_products, 'Product Prices': ser_prices}
df = pd.DataFrame(data)
df
```

```python
ser_products = pd.Series(['Product A', 'Product B', 'Product C'], index = ['A', 'B', 'C'])
ser_prices = pd.Series([22250, 16600, 12500], index = ['A', 'B', 'c'])
data = {'Product Name': ser_products, 'Product Prices': ser_prices}
df = pd.DataFrame(data)
df
```

```python
ser_products = pd.Series(['Product A', 'Product B', 'Product C'], index = ['A', 'B', 'C'])
ser_prices = pd.Series([22250, 16600, 12500], index =['C', 'A', 'B'])
data = {'Product Name': ser_products, 'Product Prices': ser_prices}
df = pd.DataFrame(data)
df
```
