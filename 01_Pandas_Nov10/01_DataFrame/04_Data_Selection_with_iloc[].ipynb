#Data Selection with .iloc[]



#pandas.DataFrame.iloc
Purely integer-location based indexinf for selection by position.

*  **iloc: i - integer, loc-location**
*  iloc[] = iloc indexer = iloc accessor


1. Same rules apply for indexing:
* Python lists
* Series by indexing
* Python Series and DataFrames with .iloc[]


 strict implicit, integer location, position-based indexing

```python
import pandas as pd
```

```python
data = pd.read_csv('Lending-company.csv', index_col='StringID')
lending_co_data = data.copy()
lending_co_data.head()
```

```python
lending_co_data[1]
```

```python
lending_co_data['DailyRate'].head()
```

```python
#row specifier = row indexer
lending_co_data.iloc[1]
```

```python
lending_co_data.iloc[1, 7]
```

```python
lending_co_data.iloc[0:5]
```

```python
#[start: end, column]
lending_co_data.iloc[:, 7]
```

```python
#[[rows we want], rows:columns]
lending_co_data.iloc[[1, 7],:]
```

```python
lending_co_data.iloc[[1, 7],:7]
```

```python
#to retrive entire columns
lending_co_data.iloc[:,[1, 7]]
```
