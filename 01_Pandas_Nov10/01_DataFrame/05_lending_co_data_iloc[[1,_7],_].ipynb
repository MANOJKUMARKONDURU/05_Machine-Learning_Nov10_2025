# .loc[]

= loc indexer = loc accessor

```python
import pandas as pd
```

```python
data = pd.read_csv('Lending-company.csv', index_col='StringID')
lending_co_data = data.copy()
lending_co_data.head()
```

```python
#to retrive everything in a specific row
lending_co_data.loc['LoanID_1']
```

```python
lending_co_data.loc['LoanID_1', 'Deposit']
```

```python
lending_co_data.loc[:,'Location']
```

```python
lending_co_data.loc[:,['Location', 'Region']].head()
```
